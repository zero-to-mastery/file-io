## Let's functionize the code

```python
def spotify_metrics(file_path):
  #read the file
  file = pd.read_csv(file_path, encoding = "latin1")

  #get no of songs
  no_of_songs = file["track_name"].nunique()
  print("Number of songs:", no_of_songs)

  #get the number of songs on key E
  file_unique = file.drop_duplicates(subset="track_name")
  num_E = (file_unique["key"]=="E").sum()
  print("Number of songs on key E:", num_E)

  #get the most common artist
  counts = file["artist(s)_name"].value_counts()
  most_common = counts.idxmax()
  frequency = counts.max()
  print("The most common artist is", most_common)
  print("The most common artist appears", frequency, "times")

spotify_metrics(file_path=file_path)

```
